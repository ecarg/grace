```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

CNN5
----

### 음절(chr), 자소(pho)

```python
cnn5_chr = pd.DataFrame.from_csv('v0.5/cnn5.chr.tsv', sep='\t')
cnn5_chr.columns = ['loss.chr', 'accuracy.chr', 'f-score.chr']
print(cnn5_chr[:5])
print()
cnn5_pho = pd.DataFrame.from_csv('v0.5/cnn5.pho.tsv', sep='\t')
cnn5_pho.columns = ['loss.pho', 'accuracy.pho', 'f-score.pho']
print(cnn5_pho[:5])
cnn5 = pd.concat([cnn5_chr, cnn5_pho])
cnn5[:10]
```

```python
cnn5.plot(y=['loss.chr', 'loss.pho'], figsize=(15, 10))
```

```python
print(max(cnn5['f-score.chr']), max(cnn5['f-score.pho']))
cnn5.plot(y=['accuracy.chr', 'accuracy.pho', 'f-score.chr', 'f-score.pho'], figsize=(15, 10))
```
