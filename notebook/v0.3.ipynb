```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

FNN
----

### 음절(chr), 자소(pho)

```python
fnn_chr = pd.DataFrame.from_csv('v0.3/fnn3.chr.tsv', sep='\t')
fnn_chr.columns = ['loss.chr', 'accuracy.chr', 'f-score.chr']
print(fnn_chr[:5])
print()
fnn_pho = pd.DataFrame.from_csv('v0.3/fnn3.pho.tsv', sep='\t')
fnn_pho.columns = ['loss.pho', 'accuracy.pho', 'f-score.pho']
print(fnn_pho[:5])
fnn = pd.concat([fnn_chr, fnn_pho])
fnn[:10]
```

```python
fnn.plot(y=['loss.chr', 'loss.pho'], figsize=(15, 10))
```

```python
print(max(fnn['f-score.chr']), max(fnn['f-score.pho']))
fnn.plot(y=['accuracy.chr', 'accuracy.pho', 'f-score.chr', 'f-score.pho'], figsize=(15, 10))
```
